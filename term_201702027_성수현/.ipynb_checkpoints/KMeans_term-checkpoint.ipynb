```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def File_Open():
    #Credit Card Dataset
    data = pd.read_csv("./CC GENERAL.csv")
   
    # PURCHASES_FREQUENCY : How frequently the Purchases are being made(구매빈도)
    # PURCHASES_INSTALLMENTS_FREQUENCY : How frequently purchases in installments are being done(할부구매빈도)
    data = data.loc[:, ['PURCHASES_FREQUENCY', 'PURCHASES_INSTALLMENTS_FREQUENCY']]

    data = np.array(data)
    return data
```

```python
def Draw_Graph(data, labels):
    plt.figure()
    plt.scatter(data[:,0],data[:,1],c=labels,cmap='rainbow')
    plt.show()
    return
```

```python


def KMeans(data,k):
    from sklearn.cluster import KMeans
    kmeans = KMeans(n_clusters= k, random_state=0).fit(data)
    labels = kmeans.labels_
    centers = kmeans.cluster_centers_
    print(centers)
    return data, labels
```

```python

if __name__ == '__main__':
    data = File_Open()
    k=5

    kmeans, labels = KMeans(data, k)
    Draw_Graph(kmeans,labels)
```
